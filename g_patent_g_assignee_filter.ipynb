```python
import pandas as pd
import os

working_dir = r"C:\\Users\\dbvan\\Downloads\\projects\\AIP 197" 
os.chdir(working_dir)
patent_cols = ["patent_id", "patent_date"]
g_patent = pd.read_csv("g_patent.tsv", sep="\t", usecols=patent_cols)

g_patent['year'] = pd.to_datetime(g_patent['patent_date'], errors='coerce').dt.year


g_patent_clean = g_patent[['patent_id', 'year']]
g_patent_clean.to_csv("g_patent_clean.csv", index=False)

assignee_cols = ["patent_id", "assignee_id", "disambig_assignee_organization", "assignee_type"]
g_assignee = pd.read_csv("g_assignee_disambiguated.tsv", sep="\t", usecols=assignee_cols)

g_assignee_companies = g_assignee[g_assignee['assignee_type'].isin([2, 3])]

g_assignee_clean = g_assignee_companies[['patent_id', 'assignee_id', 'disambig_assignee_organization']]

g_assignee_clean.to_csv("g_assignee_clean.csv", index=False)

print("Cleaning done. Cleaned files saved as g_patent_clean.csv and g_assignee_clean.csv")

```
