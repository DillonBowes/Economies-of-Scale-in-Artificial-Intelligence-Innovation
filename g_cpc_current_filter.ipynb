```python
import pandas as pd

# TSV
working_dir = "C:\\Users\\dbvan\\Downloads\\projects\\AIP 197\\" 
cpc_file = working_dir + "g_cpc_current.tsv"
cols_needed = ['patent_id', 'cpc_group']
cpc_df = pd.read_csv(cpc_file, sep='\t', usecols=cols_needed, dtype=str)

#AI-related CPC codes
ai_codes = [
    "G06N", "G06N3/00", "G06N3/02", "G06N3/04", "G06N3/08",
    "G06N3/10", "G06N3/12", "G06N3/14", "G06N5/00", "G06N7/00",
    "G06F40/00", "G10L15/00", "G06K9/00", "G06Q10/00"
]

ai_patents = cpc_df[cpc_df['cpc_group'].isin(ai_codes)]
ai_patents = ai_patents.drop_duplicates(subset=['patent_id'])
ai_patents.to_csv(working_dir + "ai_patents.csv", index=False)

print(f"Filtered {len(ai_patents)} AI patents. Saved to 'ai_patents.csv'")
```
