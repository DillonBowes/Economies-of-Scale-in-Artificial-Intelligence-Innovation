```python
import pandas as pd
import os

working_dir = r"C:\\Users\\dbvan\\Downloads\\projects\\AIP 197\\secondaries" 
os.chdir(working_dir)

g_patent = pd.read_csv("g_patent_clean.csv", dtype={"patent_id": str})
g_assignee = pd.read_csv("g_assignee_clean.csv", dtype={"patent_id": str})
g_cpc = pd.read_csv("g_cpc_clean.csv", dtype={"patent_id": str})

ai_cpc_codes = [
    "G06N",
    "G06N3/00","G06N3/02","G06N3/04","G06N3/08","G06N3/10","G06N3/12","G06N3/14",
    "G06N5/00","G06N7/00",
    "G06F40/00",
    "G10L15/00",
    "G06K9/00",
    "G06Q10/00"
]
patent_ai = g_cpc[g_cpc['cpc_group'].isin(ai_cpc_codes)].copy()
ai_with_company = pd.merge(patent_ai, g_assignee, on="patent_id", how="left")
ai_with_company = pd.merge(ai_with_company, g_patent, on="patent_id", how="left")
ai_counts = (
    ai_with_company.groupby(["disambig_assignee_organization", "year"])
    .size()
    .reset_index(name="ai_patent_count")
    .sort_values(["disambig_assignee_organization", "year"])
)

ai_counts.to_csv("ai_patent_counts_per_company_year.csv", index=False)

print("Done! CSV saved as ai_patent_counts_per_company_year.csv")
```
