```python
import os
import pandas as pd

working_dir = "C:\\Users\\dbvan\\Downloads\\projects\\AIP 197\\tertiaries"
os.chdir(working_dir)

ai_counts = pd.read_csv("ai_patent_counts_per_company_year.csv")

ai_counts = ai_counts.rename(columns={
    "disambig_assignee_organization": "conm",
    "year": "fyear"
})

def clean_name(s):
    return (
        s.astype(str)
         .str.upper()
         .str.replace(r"[.,/]", "", regex=True)
         .str.replace(r"\b(INC|CORP|CORPORATION|LLC|LTD|L L C)\b", "", regex=True)
         .str.replace(r"\s+", " ", regex=True)
         .str.strip()
    )

ai_counts["conm"] = clean_name(ai_counts["conm"])
ai_compustat_fmt = (
    ai_counts
        .groupby(["conm", "fyear"], as_index=False)
        .agg(ai_patent_count=("ai_patent_count", "sum"))
)

ai_compustat_fmt = ai_compustat_fmt.sort_values(
    by=["conm", "fyear"]
)
output_file = "ai_patents_compustat_format.csv"
ai_compustat_fmt.to_csv(output_file, index=False)

print(f"Saved file: {output_file}")
print(f"Total firm-year rows: {len(ai_compustat_fmt)}")
```
