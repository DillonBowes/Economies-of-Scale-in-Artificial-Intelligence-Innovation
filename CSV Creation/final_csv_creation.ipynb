```python
import os
import pandas as pd

working_dir = "C:\\Users\\dbvan\\Downloads\\projects\\AIP 197\\tertiaries"
os.chdir(working_dir)

comp = pd.read_csv("compustat_data.csv")
ai = pd.read_csv("ai_patents_compustat_format.csv")
def clean_name(s):
    return (
        s.astype(str)
         .str.upper()
         .str.replace(r"[.,/]", "", regex=True)
         .str.replace(r"\b(INC|CORP|CORPORATION|LLC|LTD|L L C)\b", "", regex=True)
         .str.replace(r"\s+", " ", regex=True)
         .str.strip()
    )

comp["conm"] = clean_name(comp["conm"])
ai["conm"] = clean_name(ai["conm"])

merged = comp.merge(
    ai,
    how="inner",
    on=["conm", "fyear"]
)
output_file = "compustat_with_ai_patents_inner_join.csv"
merged.to_csv(output_file, index=False)

print(f"Saved merged file: {output_file}")
print(f"Matched firm-year observations: {len(merged)}")
print(f"Unique firms: {merged['conm'].nunique()}")
```
